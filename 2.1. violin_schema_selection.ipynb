```python
import pandas as pd


# Load as DataFrames
# File paths
file_vaccine = "Dataset/VIOLIN_12-10-2025/raw/t_vaccine_202512101310.csv"
file_adjuvant = "Dataset/VIOLIN_12-10-2025/raw/t_adjuvant_202512101312.csv"
file_vaxjo   = "Dataset/VIOLIN_12-10-2025/raw/t_vaxjo_202512101311.csv"
file_pathogen   = "Dataset/VIOLIN_12-10-2025/raw/t_pathogen_202512101313.csv"
file_reference   = "Dataset/VIOLIN_12-10-2025/raw/t_reference_202512101313.csv"

# Load as DataFrames
df_vaccine  = pd.read_csv(file_vaccine, encoding="utf-8-sig") #encoding="utf-8"  # this param is ignored for .xlsx but good practice #encoding='ISO-8859-1', on_bad_lines='warn'
df_adjuvant = pd.read_csv(file_adjuvant, encoding="utf-8-sig")
df_vaxjo    = pd.read_csv(file_vaxjo, encoding="utf-8-sig")
df_pathogen = pd.read_csv(file_pathogen, encoding="utf-8-sig")

df_final = pd.read_csv("Dataset/VIOLIN_12-10-2025/interim/infectious_df_pubmed_final.csv", encoding="utf-8-sig")

# Count columns
column_counts = {
    "t_vaxjo": len(df_vaxjo.columns),
    "t_adjuvant": len(df_adjuvant.columns),
    "t_vaccine": len(df_vaccine.columns),
    "t_pathogen": len(df_pathogen.columns),
    "infectious_df_pubmed_final": len(df_final.columns)
}

# Print column counts
print("Column counts:")
for name, count in column_counts.items():
    print(f"{name}: {count} columns")
print()

# Build origin lookup map
origin_map = {}

# Helper function to track columns
def track_columns(df, source_name):
    for col in df.columns:
        origin_map[col] = source_name

# Track original source columns
track_columns(df_vaxjo, "t_vaxjo")
track_columns(df_adjuvant, "t_adjuvant")
track_columns(df_vaccine, "t_vaccine")
track_columns(df_pathogen, "t_pathogen")

# Map columns from final file
final_column_sources = []

for col in df_final.columns:
    # Try exact match
    if col in origin_map:
        source = origin_map[col]
    else:
        # Try by suffix
        for suffix, source_guess in {
            "_vaxjo": "t_vaxjo",
            "_adjuvant": "t_adjuvant",
            "_vaccine": "t_vaccine",
            "_pathogen": "t_pathogen"
        }.items():
            if col.endswith(suffix):
                source = source_guess
                break
        else:
            source = "unknown"
    final_column_sources.append((col, source))

# Print source mapping
print(f"{'Column Name':<50} | {'Source Table'}")
print("-" * 70)
for col, source in final_column_sources:
    print(f"{col:<50} | {source}")
```
