```python
from IPython.display import display

def show_test_rows_full(df, col, n=10):
    if col in df.columns:
        mask = df[col].astype(str).str.lower().str.startswith("test")
        print(f"{col}: {mask.sum()} test rows")
        return df.loc[mask].head(n)
    return None

display(show_test_rows_full(df_adjuvant_raw, "c_adjuvant_label"))
display(show_test_rows_full(df_vaxjo_raw, "c_vaxjo_name"))
display(show_test_rows_full(df_vaccine_raw, "c_vaccine_name"))
display(show_test_rows_full(df_pathogen_raw, "c_pathogen_name"))
```

```python
def drop_test_rows(df, col):
    if col in df.columns:
        mask = df[col].astype(str).str.lower().str.startswith("test")
        return df[~mask]
    return df

df_adjuvant = drop_test_rows(df_adjuvant_raw, "c_adjuvant_label")
df_vaxjo   = drop_test_rows(df_vaxjo_raw, "c_vaxjo_name")
df_vaccine = drop_test_rows(df_vaccine_raw, "c_vaccine_name")
df_pathogen = drop_test_rows(df_pathogen_raw, "c_pathogen_name")

df_adjuvant.to_csv("Dataset/VIOLIN_12-10-2025/raw/t_adjuvant_202512101312_clean.csv", index=False, encoding="utf-8-sig")
df_vaxjo.to_csv("Dataset/VIOLIN_12-10-2025/raw/t_vaxjo_202512101311_clean.csv", index=False, encoding="utf-8-sig")
df_vaccine.to_csv("Dataset/VIOLIN_12-10-2025/raw/t_vaccine_202512101310_clean.csv", index=False, encoding="utf-8-sig")
df_pathogen.to_csv("Dataset/VIOLIN_12-10-2025/raw/t_pathogen_202512101313_clean.csv", index=False, encoding="utf-8-sig")

print("Saved cleaned files to raw/")
# Summaries
summarize_df(df_vaccine, "t_vaccine")
summarize_df(df_adjuvant, "t_adjuvant")
summarize_df(df_vaxjo, "t_vaxjo")
summarize_df(df_pathogen, "t_pathogen")
```

```python
def drop_identical_suffixed_columns(df, suffix):
    dropped = []
    for col in df.columns:
        if col.endswith(suffix):
            base_col = col.replace(suffix, '')
            if base_col in df.columns:
                same = df[col].fillna("__") == df[base_col].fillna("__")
                if same.all():
                    df.drop(columns=[col], inplace=True)
                    dropped.append(col)
    return dropped

# Example use:
dropped_cols = drop_identical_suffixed_columns(df_adjuvant_ext1, suffix = "_vaxjo")
print("Dropped columns:", dropped_cols)
df_adjuvant_ext1.head(2)
```

```python
def analyze_remaining_suffixed_columns(df, suffix, show_table=True):
    """
    Analyze remaining column pairs after dropping identicals.
    Compares base vs. suffixed column and returns match statistics.

    Args:
        df (pd.DataFrame): Input DataFrame
        suffix (str): Suffix to look for (e.g., "_vaxjo")
        show_table (bool): Whether to display the table

    Returns:
        pd.DataFrame: Summary comparison table
    """
    import pandas as pd

    results = []

    col_pairs = [
        (col.replace(suffix, ''), col)
        for col in df.columns
        if col.endswith(suffix) and col.replace(suffix, '') in df.columns
    ]

    for base_col, suffixed_col in col_pairs:
        base = df[base_col]
        comp = df[suffixed_col]

        both_empty = base.isna() & comp.isna()
        same_non_empty = (base == comp) & ~both_empty
        different = (base != comp) & ~(base.isna() | comp.isna())
        left_empty_only = base.isna() & ~comp.isna()
        right_empty_only = ~base.isna() & comp.isna()

        results.append({
            "Column": base_col,
            "Same (non-empty)": same_non_empty.sum(),
            "Different": different.sum(),
            "Both empty": both_empty.sum(),
            "Original empty only": left_empty_only.sum(),
            "Suffixed empty only": right_empty_only.sum()
        })

    summary_df = pd.DataFrame(results)
    summary_df = summary_df[[
        "Column", "Same (non-empty)", "Different",
        "Both empty", "Original empty only", "Suffixed empty only"
    ]].sort_values(by="Column").reset_index(drop=True)


    return summary_df
```

```python
import pandas as pd
import re

# Load your DataFrame
df = infectious_df

# Define regex pattern for [Ref...] or [PubMed...]
pattern = r'\[(Ref\d+:[^\]]+|PubMed:[^\]]+)\]'

# Dictionary to store results
column_ref_counts = {}

# Loop through each column
for col in df.columns:
    # Drop NaNs and convert all entries to string, then join into one large string
    combined_text = ' '.join(df[col].dropna().astype(str))
    
    # Find all matches of the pattern
    matches = re.findall(pattern, combined_text)
    
    # Count unique matches
    unique_refs = set(matches)
    
    # Store the count
    column_ref_counts[col] = len(unique_refs)

# Convert result to a DataFrame for easy viewing (optional)
ref_counts_df = pd.DataFrame.from_dict(column_ref_counts, orient='index', columns=['unique_reference_count'])
ref_counts_df = ref_counts_df.sort_values(by='unique_reference_count', ascending=False)

# Show the result
print(ref_counts_df.head(50))
```

```python
import pandas as pd
import re


# Patterns
ref_pattern = r'\[(Ref\d+:[^\]]+)\]'
pubmed_pattern = r'\[(PubMed:[^\]]+)\]'

# Storage
stats = []

# Loop through each column
for col in df.columns:
    combined_text = ' '.join(df[col].dropna().astype(str))

    # Extract matches
    ref_matches = re.findall(ref_pattern, combined_text)
    pubmed_matches = re.findall(pubmed_pattern, combined_text)

    # Convert to sets to ensure uniqueness
    unique_ref = set(ref_matches)
    unique_pubmed = set(pubmed_matches)

    # Combine for total
    total_unique = unique_ref.union(unique_pubmed)

    # Store the result
    stats.append({
        'column': col,
        'total_unique_refs': len(total_unique),
        'ref_count': len(unique_ref),
        'pubmed_count': len(unique_pubmed),
        'has_ref': len(unique_ref) > 0,
        'has_pubmed': len(unique_pubmed) > 0
    })

# Convert to DataFrame for better display
ref_stats_df = pd.DataFrame(stats).sort_values(by='total_unique_refs', ascending=False)

# Show results
print(ref_stats_df.head(40))
```
